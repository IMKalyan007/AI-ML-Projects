```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import math
import numpy as np
```

```python

cd_train=pd.read_csv('Consumer_Complaints_train.csv')
cd_test=pd.read_csv('Consumer_Complaints_test_share.csv')
```

```python
cd_train['ZIP code']
```

```python
cd_train.shape
```

```python
for col in ['Date received','Date sent to company']:
    cd_train[col]=pd.to_datetime(cd_train[col],infer_datetime_format=True)
    cd_test[col]=pd.to_datetime(cd_test[col],infer_datetime_format=True)
```

```python
cd_train['day_diff']=pd.to_numeric(cd_train['Date sent to company']-cd_train['Date received'])
cd_test['day_diff']=pd.to_numeric(cd_test['Date sent to company']-cd_test['Date received'])
```

```python
for col in ['Date received','Date sent to company']:
    cd_train.drop([col],1,inplace=True)
    cd_test.drop([col],1,inplace=True)
```

```python
len(pd.isnull(cd_train['Tags']))
len(cd_train)
```

```python
for col in ['Sub-product','Sub-issue','Consumer complaint narrative',
            'Company public response','Tags','Consumer consent provided?']:
    varname=col.replace('-','_').replace('?','').replace(" ",'_')+'_isNan'
    cd_train[varname]=np.where(pd.isnull(cd_train[col]),1,0)
    cd_train.drop([col],1,inplace=True)
    cd_test[varname]=np.where(pd.isnull(cd_test[col]),1,0)
    cd_test.drop([col],1,inplace=True)
    
```

```python
cd_train['Consumer disputed?']=np.where(cd_train['Consumer disputed?']=="Yes",1,0)
```

```python
k=cd_train['Issue'].value_counts()
for val in k.axes[0][0:10]:
    varname='Issue_'+val.replace(',','_').replace(' ','_')
    cd_train[varname]=np.where(cd_train['Issue']==val,1,0)
    cd_test[varname]=np.where(cd_test['Issue']==val,1,0)
del cd_train['Issue']
del cd_test['Issue']
```

```python
for col in ['Product','Submitted via','Company response to consumer','Timely response?']:
    
    temp=pd.get_dummies(cd_train[col],prefix=col,drop_first=True)
    cd_train=pd.concat([temp,cd_train],1)
    cd_train.drop([col],1,inplace=True)
    
    temp=pd.get_dummies(cd_test[col],prefix=col,drop_first=True)
    cd_test=pd.concat([temp,cd_test],1)
    cd_test.drop([col],1,inplace=True)
```

```python
k=cd_train['State'].value_counts()
for val in k.axes[0][0:15]:
    varname='State_'+val.replace(',','_').replace(' ','_')
    cd_train[varname]=np.where(cd_train['State']==val,1,0)
    cd_test[varname]=np.where(cd_test['State']==val,1,0)
del cd_train['State']
del cd_test['State']
```

```python
for col in ['Timely response?_Yes', 'Company response to consumer_Closed with explanation','Company response to consumer_Closed with monetary relief',
           'Company response to consumer_Closed with non-monetary relief','Company response to consumer_Closed with relief','Company response to consumer_Closed without relief',
           'Company response to consumer_Untimely response',
           'Submitted via_Fax','Submitted via_Phone','Submitted via_Postal mail','Submitted via_Referral', 'Submitted via_Web',
           'Product_Consumer Loan','Product_Credit card','Product_Credit reporting','Product_Debt collection',                                         
           'Product_Money transfers','Product_Mortgage','Product_Other financial service','Product_Payday loan',                                              
           'Product_Prepaid card','Product_Student loan','Product_Virtual currency']:
    cd_train[col] =np.int32(cd_train[col])
```

```python
for col in ['Timely response?_Yes', 'Company response to consumer_Closed with explanation','Company response to consumer_Closed with monetary relief',
           'Company response to consumer_Closed with non-monetary relief','Company response to consumer_Closed with relief','Company response to consumer_Closed without relief',
           'Company response to consumer_Untimely response',
           'Submitted via_Fax','Submitted via_Phone','Submitted via_Postal mail','Submitted via_Referral', 'Submitted via_Web',
           'Product_Consumer Loan','Product_Credit card','Product_Credit reporting','Product_Debt collection',                                         
           'Product_Money transfers','Product_Mortgage','Product_Other financial service','Product_Payday loan',                                              
           'Product_Prepaid card','Product_Student loan','Product_Virtual currency']:
    cd_test[col] =np.int32(cd_test[col])
```

```python
cd_train['ZIP code'] = cd_train['ZIP code'].str.replace('XX',"00")
cd_train['ZIP code']=pd.to_numeric(cd_train['ZIP code'],errors='coerce')
```

```python
cd_test['ZIP code'] = cd_test['ZIP code'].str.replace('XX',"00")
cd_test['ZIP code']=pd.to_numeric(cd_test['ZIP code'],errors='coerce')
```

```python
cd_train.shape
```

```python
cd_test.shape
```

```python
cd_train['Company'] = cd_train['Company'].str.replace(' ','')
cd_train['Company'] = cd_train['Company'].str.replace(',','')
cd_train['Company'] = cd_train['Company'].str.replace('&','')
cd_train['Company'] = cd_train['Company'].str.replace('.','')

cd_test['Company'] = cd_test['Company'].str.replace(' ','')
cd_test['Company'] = cd_test['Company'].str.replace(',','')
cd_test['Company'] = cd_test['Company'].str.replace('&','')
cd_test['Company'] = cd_test['Company'].str.replace('.','')
```

```python
cat_cols=cd_train['Company']

to_store = []
new_cols = []

#for col in cat_cols:
    
freqs = cd_train['Company'].value_counts()
categories = freqs.index[freqs>2000]
to_store.append(('Company', categories))
    
for category in categories:
    name=col+'.'+category
    new_cols.append(name)
    cd_train[name]=(cd_train['Company']==category).astype(int)
del cd_train['Company']
#print(categories)





```

```python
for column, categories in to_store:
        for category in categories:
            name=column+'.'+category
            cd_test[name]=(cd_test[column]==category).astype(int)
        del cd_test[column]
```

```python
# Handling missing value
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
cd_train = pd.DataFrame(imputer.fit_transform(cd_train), columns=cd_train.columns)

cd_test = pd.DataFrame(imputer.fit_transform(cd_test), columns=cd_test.columns)
```

```python
#Split the data set for train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cd_train.drop(['Consumer disputed?','Complaint ID'], axis=1), 
                                                    cd_train['Consumer disputed?'], test_size=0.2, random_state=123)

print('Train values shape:', X_train.shape)
print('Test values shape:', X_test.shape)
print('Train target shape:', y_train.shape)
print('Test target shape:', y_test.shape)





xTestSub=cd_test['Complaint ID']

cd_test=cd_test.drop(['Complaint ID'],1)
```

```python
cd_test.shape
```

```python
# scalling the input data

from sklearn.preprocessing import StandardScaler


sc_X = StandardScaler() 
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)

cd_test=sc_X.fit_transform(cd_test)

# Normalizing continuous variables

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))

scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

cd_test=scaler.fit_transform(cd_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python

from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score

modelLR = LogisticRegression(solver='liblinear',random_state=123)
modelLR.fit(X_train,y_train)

y_train_hat=modelLR.predict(X_train)
y_test_hat=modelLR.predict(X_test)
y_test_hat_probs = modelLR.predict_proba(X_test)[:,1]


print(modelLR)
print('Train performance')
print('-------------------------------------------------------')
print(classification_report(y_train,y_train_hat))

print('Test performance')
print('-------------------------------------------------------')
print(classification_report(y_test,y_test_hat))

print('roc_auc_score')
print('-------------------------------------------------------')
print(roc_auc_score(y_test,y_test_hat))

print('confusion_matrix')
print('-------------------------------------------------------')
print(confusion_matrix(y_test,y_test_hat))
```
